# BeamCurrentAnalyzer - Demo

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import hyperspy.api as hs
import uts        # Importing the uts.py module
import importlib
```

## Loading data and initializing the BeamCurrentAnalyzer class

```python
params = {
    'folder_current': r'.\data_spot_analyzer\beam_current', # Path to the folder containing the .dm3 files.
    'format': 'current{d}',    # Format of .dm3 files. Optional.
    'counts_per_electron': 1.47,    # Specify the conversion efficiency
}
a = uts.BeamCurrentAnalyzer(params)
```

## Analyzing the data

List the beam currents

```python
a.get_beam_current()
```

Plot beam currents

```python
# Each blue spot is the beam current measured at the particular spot. Note that each spot is based on ten individual measurements. The spot is the median of the ten measurements.

a.plot_beam_current()
```

```python
# The same plot as above, shown at greater detail. The green spots indicate the ten individual measurements. The blue spot is chosen as the median of the ten measurements. Note that the x position of each measurements corresponds exactly to the time at which the measurement was done.

a.plot_beam_current(full=True)
plt.xlim(3.097, 3.112)
plt.ylim(0.03505, 0.03537)
```
